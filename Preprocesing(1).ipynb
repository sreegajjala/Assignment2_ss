```python
import pandas as pd
```

```python
car_data=pd.read_csv("/content/train.csv")
```

```python
car_data.head()
```

```python
#Check for null values
```

```python
car_data.isna().sum()
```

```python
car_data2 = pd.DataFrame(car_data)
```

```python
# (b)
```

```python
car_data2['Mileage'] = car_data2['Mileage'].str.extract('(\d+\.\d+)').astype(float)

car_data2['Engine'] = car_data2['Engine'].str.replace(' CC', '').astype(float)

car_data2['Power'] = car_data2['Power'].str.extract('(\d+\.\d+)').astype(float)

car_data2['New_Price'] = car_data2['New_Price'].str.extract('(\d+\.\d+)').astype(float)
```

```python
# (a)
```

```python
car_data2['Mileage'].fillna(car_data2['Mileage'].median(),inplace=True)
car_data2['Engine'].fillna(car_data2['Engine'].median(),inplace=True)
car_data2['Power'].fillna(car_data2['Power'].median(),inplace=True)

car_data2['Seats'].fillna(car_data2['Seats'].mode()[0],inplace=True)

car_data2.drop('New_Price', axis=1, inplace=True)#axis=1 indicates to drop a column
```

```python
output_csv_file = 'cleaned_data.csv'
car_data2.to_csv(output_csv_file, index=False)
```
